```python
#Determinar la cantidad a pagar por el cliente en un supermercado, comprando un producto X que tiene los siguientes descuento.
#De 10  libras, a 12 libras 5% descuento
#De 12.1 a 15 libras, 10% de descuento
#De 15.1 a 20  libras, 15% descuento
print("^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^")
print("       Factura de Comercial Elia          ")
print("******************************************")

libr = float(input("cuantas libras tiene "))
prec = float(input("precio del producto: "))
desc = 0

if libr < 10 :  
  print("no cumple con la libras necesarias, es de 10 libras en adelante para recibir descuento")
elif libr >= 10 and libr <= 12 :
  print("recibira un 5%") 
  desc = 5
elif libr >= 12.1 and libr <= 15 :
  print("recibira un 10%")
  desc = 10
elif libr >= 15.1 and libr <= 20 :
  print("recibira un 15%")
  desc = 15
else : 
  print("recibera el descuento maximo que es un 15%")
  desc = 15

sub_pre = prec * libr
desc_arel = sub_pre / 100
desc_apli = desc_arel * desc
pagar = sub_pre - desc_apli
print("precio a pagar es ", pagar)
```
